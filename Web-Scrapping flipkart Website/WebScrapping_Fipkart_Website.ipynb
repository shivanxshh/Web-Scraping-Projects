```python
names=[]
prices=[]
ratings=[]
reviews=[]
for i in range(len(mobiles)):
  mobile=mobiles[i]
  name=mobile.find("div",class_="_4rR01T").text
  price=mobile.find("div",class_="_30jeq3 _1_WHN1").text
  rating=mobile.find("div",class_="_3LWZlK").text
  review=mobile.find("span",class_="_2_R_DZ").text
  
  
  names.append(name)
  prices.append(price)
  ratings.append(rating)
  reviews.append(review)
```

```python
ratings
```

```python
names
```

```python
import pandas as pd
data={"Name":names,"Price":prices,"Rating":ratings,"Review":reviews}
#column=["Name","Price","Rating","Reviews"]
#values=[[names],[prices],[ratings],[reviews]]
#
df=pd.DataFrame(data)
df
```
