```python
from selenium import webdriver
def web_driver():
    options = webdriver.ChromeOptions()
    options.add_argument("--verbose")
    options.add_argument('--no-sandbox')
    options.add_argument('--headless')
    options.add_argument('--disable-gpu')
    options.add_argument("--window-size=1920, 1200")
    options.add_argument('--disable-dev-shm-usage')
    driver = webdriver.Chrome(options=options)
    return driver
```

```python
driver=web_driver()
```

```python
import time
driver.get("https://summerofcode.withgoogle.com/archive/2021/organizations")
time.sleep(2)
```

```python
page=driver.page_source
```

```python
page
```

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(page, "html.parser")
```

```python
names=[]
technologies=[]
topicsused=[]

for org in orgs:
  org_name=org.find("div",class_="name").text
  url=org.a['href']
  realurl='https://summerofcode.withgoogle.com/'+url
  driver.get(realurl)
  time.sleep(2)
  page1=driver.page_source
  soup1=BeautifulSoup(page1,"html.parser")
  techno=soup1.find("div",class_="tech__content")
  techs=techno.text.split(',')
  topics=soup1.find("div",class_="topics__content")
  tops=topics.text.split(',')

  names.append(org_name)
  technologies.append(techs)
  topicsused.append(tops)





```

```python
names
```

```python
topicsused
```

```python
import pandas as pd
data={"Name of organizations":names,"Technologies in Use":technologies,"Topics":topicsused}
df=pd.DataFrame(data)
df
```

```python
csv_data=df.to_csv()
```
